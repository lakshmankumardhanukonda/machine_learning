```python
data.drop("Geography",axis=1,inplace=True)
```

```python
s=pd.get_dummies(data["Gender"],drop_first=1)
```

```python
data.head()
```

```python
result=pd.concat([data,g,s],axis=1)
```

```python
result.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(result.drop('Exited',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
result['Cluster'] = result['Exited']
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(confusion_matrix(result['Cluster'],kmeans.labels_))
print(classification_report(result['Cluster'],kmeans.labels_))
print(accuracy_score(result['Cluster'],kmeans.labels_))
```
